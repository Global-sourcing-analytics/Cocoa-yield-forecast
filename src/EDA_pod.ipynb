POL: Dane są zbierane z różnych sektorów->pod-sektorów->drzew, tak więc na jeden miesiąc przypada wiele punktów co nie jest czymś pożądanym. Z tego względu z każdego miesiąca powinna zostać wyciągnięta suma zbiorów. Dlaczego suma? Wynika to z specyfikacji danych. W tym przypadku średnia oznaczałaby średnią wartość zbioru z jednego drzewa, natomiast nas interesuje suma zbioru w kraju z danego miesiąca.

ENG: The data are collected from different different sectors->sub-sectors->trees, so there are multiple points per month which is not something desirable. Therefore, a harvest total should be drawn from each month. Why a total? This is due to the specification of the data. In this case, average would mean the average value of the harvest from one tree, whereas we are interested in the total harvest in the country from a given month.

Otrzymane szeregi czasowe nie są ciągłe w czasie, interwał czasowy nie wynosi zawsze miesiąc. Z tego względu należy przeprowadzić interpolację.

```python
df_monthly.set_index('Date', inplace=True)
```

```python
df_monthly_interpolated = pd.DataFrame()
df_monthly_interpolated
```

```python
df_monthly_interpolated = pd.DataFrame()
for country in df_monthly["Country"].unique():
    period_array = pd.date_range(df_monthly[df_monthly["Country"] == country].head(1).index.values[0], df_monthly[df_monthly["Country"] == country].tail(1).index.values[0] + pd.DateOffset(months=1), freq='M')
    period_array = [i.replace(day=1).date() for i in period_array]
    df_full_period = pd.DataFrame(period_array)
    df_full_period = df_full_period.rename(columns={0: 'Date'})
    df_full_period["Value"] = 0
    df_full_period.set_index("Date", inplace=True)
    df_monthly_new = pd.merge(df_monthly[df_monthly["Country"] == country], df_full_period, right_index=True, left_index=True, how='right')
    df_monthly_new = df_monthly_new.drop(columns=['Value'])
    df_monthly_new["Country"] = df_monthly_new["Country"].fillna(country)
    df_monthly_new.plot(title=country)
    df_monthly_new["Total"] = df_monthly_new["Total"].interpolate()
    print(df_monthly_new)
    df_monthly_new.plot(title=country)
    df_monthly_interpolated = df_monthly_interpolated.append(df_monthly_new)


```

```python
df_monthly_interpolated
```

## Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
for country in df_monthly_interpolated["Country"].unique():
    decomp = seasonal_decompose(df_monthly_interpolated[df_monthly_interpolated["Country"] == country].drop(columns=["Country"]))
    # Plot the decomposition
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 8))
    decomp.observed.plot(ax=ax1)
    ax1.set_ylabel('Observed')
    decomp.trend.plot(ax=ax2)
    ax2.set_ylabel('Trend')
    decomp.seasonal.plot(ax=ax3)
    ax3.set_ylabel('Seasonal')
    decomp.resid.plot(ax=ax4)
    ax4.set_ylabel('Residual')

    # Add a title to the plot
    fig.suptitle(country)

    # Show the plot
    plt.show()
```

Wszystkie wykresy cechują się silną sezonowością co może być podstawą do hipotezy, że model SARIMA będzie wykazywał lepsze wyniki niż ARIMA. Cote d Ivoire oraz Ghana cechują się wzrostowym trendem, natomiast Cameroon oraz Brazylia trendem spadkowym. Sezonowość może być dobrze wyjaśniona przez naturę zbiorów jakichkolwiek upraw. Uprawy muszą rosnąć, więc w szczycie zbiorów będzie ich najwięcej, podczas gdy poza sezonem będzie ich bardzo mało.

## Stationarity: Dickey-Fuller test

```python
from statsmodels.tsa.stattools import adfuller
```

```python
for country in df_monthly_interpolated["Country"].unique():
    adf, pval, usedlag, nobs, crit_vals, icbest =  adfuller(df_monthly_interpolated[df_monthly_interpolated["Country"] == country].drop(columns=["Country"]).Total.values)
    print("********************")
    print(country)
    print('ADF test statistic:', adf)
    print('ADF p-values:', pval)
    print('ADF number of lags used:', usedlag)
    print('ADF number of observations:', nobs)
    print('ADF critical values:', crit_vals)
    print('ADF best information criterion:', icbest)
```

POL: p-wartość jest powyżej 0.05 a więc nie można przyjąć hipotezy alternatywnej o stacjonarności. Szeregi czasowe są więc nie-stacjonarne. Najbliżej stacjonarności jest szereg czasowy dla danych Brazylii - około 6% wartości p.

ENG: The p-value is above 0.05 and so the alternative hypothesis of stationarity cannot be accepted. The time series are therefore non-stationary.
